## Sorting User Interactions Data

```python
def find_total_event(users_interactions_row):
    total_likes = users_interactions[(users_interactions["contentId"] == users_interactions_row["contentId"]) & (users_interactions["eventType"] == "LIKE")].shape[0]
    total_views = users_interactions[(users_interactions["contentId"] == users_interactions_row["contentId"]) & (users_interactions["eventType"] == "VIEW")].shape[0]
    total_bookmarks = users_interactions[(users_interactions["contentId"] == users_interactions_row["contentId"]) & (users_interactions["eventType"] == "BOOKMARK")].shape[0]
    total_follows = users_interactions[(users_interactions["contentId"] == users_interactions_row["contentId"]) & (users_interactions["eventType"] == "FOLLOW")].shape[0]
    total_comments = users_interactions[(users_interactions["contentId"] == users_interactions_row["contentId"]) & (users_interactions["eventType"] == "COMMENT CREATED")].shape[0]
    
    return total_likes + total_views + total_bookmarks + total_comments + total_follows

users_interactions["total_events"] = users_interactions.apply(find_total_event, axis=1)
users_interactions = users_interactions.sort_values("total_events", ascending=False)
users_interactions.head()
```

# Project C140

```python
def convert_lowercase(x):
    if isinstance(x, str):
        return x.lower()
    else:
        return ''

shared_articles["title"] = shared_articles["title"].apply(convert_lowercase)
shared_articles.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(shared_articles["title"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
shared_articles = shared_articles.reset_index(drop=True)
indices = pandas.Series(shared_articles.index, index=shared_articles['contentId'])
```

```python
def get_recomendations(contentId, cosine_sim):
    idx = indices[contentId]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return shared_articles["contentId"].iloc[movie_indices]
```

```python
recomendation = get_recomendations(-4110354420726924665, cosine_sim)
print(f"You should also try watching these: \n{recomendation}")
```
